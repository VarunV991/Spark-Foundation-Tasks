```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import datasets

%matplotlib inline
```

```python
# Import Dataset

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
X[1]
```

```python
print(X.shape)
print(y.shape)
```

## Decision Tree

```python
# Initializing the decision tree

from sklearn import tree

dt = tree.DecisionTreeClassifier()
dt.fit(X,y)
```

```python
# Visualization using plot_tree method present in sklearn library

fig, axes = plt.subplots(nrows = 1,ncols =1,figsize = (11,11), dpi=200)
_ = tree.plot_tree(dt,feature_names=iris.feature_names,class_names=iris.target_names,filled=True,fontsize=7)
```

```python
# Save the figure

fig.savefig("Decision_Tree.png")
```

## Output Prediciton

```python
# Class Labels
dict_target = {0:"setosa",1:"versicolor",2:"virginica"}
```

```python
# Sample Input
inputarr = np.array([7.6,3.8,2.9,2.6]).reshape(1,-1)
```

```python
# Predicted Output
pred_output = dt.predict(inputarr)[0]
```

```python
print(f'''The given input is : {inputarr[0]}''')
print(f'''The predicted output is : {dict_target[pred_output]}''')
```
