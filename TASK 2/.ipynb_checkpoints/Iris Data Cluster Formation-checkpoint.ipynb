```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset

df = pd.read_csv("Iris.csv")
```

```python
df.head()
```

```python
df.shape
```

## Exploratory Data Analysis

```python
# Plotting different attributes based on their species

features = ["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]

for feature in features:
    sns.catplot(x = feature,y="Species",data=df)
    plt.show()
```

#### We can see that the data can be split into clusters and each cluster will have differentiating value for each attribute.

## Applying K-Means for clustering

```python
# The independent attibutes

X = df.iloc[:,[1,2,3,4]]
```

```python
# Testing different no. of clusters to get the opitmal values

from sklearn.cluster import KMeans

score = []
for i in range(1,10):
    kmeans_model = KMeans(i,n_jobs=-1)
    kmeans_model.fit(X)
    score.append(kmeans_model.inertia_)
```

```python
# J-score is used to evaluate the kmeans model. A lesses value means a better performing model.
score
```

```python
# Plotting the J-score vs no. of clusters

plt.figure(figsize=(6,6))
plt.plot(range(1,10),score,'go-')
plt.xlabel("No. of Clusters")
plt.ylabel("Within-Cluster sum of squares")
plt.title("Elbow Method to determine the optimal value of cluster size")
```

#### Here, we can see that the elbow occurs at no. of clusters = 3 . So, the optimal cluster size for the given dataset is 3. As increasing the number of clusters, will just overfit the data and there is no significant reduction in J - score. 

## Visualizing the Clusters

```python
# Creating clusters of size 3

kmeans = KMeans(3,n_jobs=-1,init='k-means++', n_init=10, max_iter=300)
```

```python
# Predicting the clusters

pred = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(8,6))

# Plot the point acc. to the clusters
plt.scatter(X[pred==0].iloc[:,0], X[pred==0].iloc[:,1], s = 50, label = 'Iris-Setosa')
plt.scatter(X[pred==1].iloc[:,0], X[pred==1].iloc[:,1], s = 50, label = 'Iris-Versicolor')
plt.scatter(X[pred==2].iloc[:,0], X[pred==2].iloc[:,1], s = 50, label = 'Iris-Virginica')

# Plot the cluster centroids
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
plt.xlabel("Sepal Length (Cm)")
plt.ylabel("Sepal Width (Cm)")
plt.legend()
```
